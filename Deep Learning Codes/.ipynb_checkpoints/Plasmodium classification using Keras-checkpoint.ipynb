# Plasmodium detection in cells using Deep Learning

```python
# Importing Keras dependency

import os
import time
from PIL import Image
import numpy as np
import pandas as pd
import cv2
import keras 
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.models import Sequential
from keras.utils import np_utils
import matplotlib.pyplot as plt
```

Higher pigmentation is seen in the cell infected with Malaria parasite

```python
# Read all images

Data = []
Outcomes = []

path  = "/Users/abhinavmathur/Downloads/cell_images"
folders = os.listdir(path)
print(folders)
```

```python
# We only need the folders

folders.pop(0)
print(folders)
```

```python
# Now write a loop to read all files and generate their labels

start_time = time.time()

for types in folders:
    result = folders.index(types)
    #print(result)
    #print(path+'/'+types)
    files = os.listdir(path+'/'+types)
    for file in files:
        try:
            img = cv2.imread(path+'/'+types+'/'+file)
            Img = Image.fromarray(img, 'RGB')
            Img2 = Img.resize((50,50))
            Data.append(np.array(Img2))
            Outcomes.append(result)
        except Exception:
            pass
print("--- %s seconds taken to load data---" % (time.time() - start_time))
```

```python
print(str(len(Data))+' are the total number of images in data')
```

```python
print(str(len(np.unique(Outcomes)))+' are the total number of classes in the data with '+str(len(Outcomes))+' number of records')
```

```python
images = np.array(Data) # Converting to a numpy array
infections = np.array(Outcomes) # Converting to a numpy array
```

```python
images.shape #Get the shape of the array
```

We need to convert the array into a single observation for a single data

```python
indexes=np.arange(images.shape[0]) #Get the total count of rows in an array
np.random.shuffle(indexes) # Randomly shuffle row numbers
images=images[indexes] # Rearrange rows in the images data
infections=infections[indexes] # Rearrange outcomes
```

```python
# Split the data into train and test datasets by taking a 80:20 split

(X_train, X_test) = images[(int)(0.7*len(images)):],images[:(int)(0.7*len(images))]
(Y_train, Y_test) = infections[(int)(0.7*len(infections)):],infections[:(int)(0.7*len(infections))]
print(len(X_train),len(X_test),len(images))
```

```python
# Normalize the data by dividing with 255 pixels
X_train = X_train.astype('float32')/255 # As we are working on image data we are normalizing data by divinding 255.
X_test = X_test.astype('float32')/255
```

```python
# As this is a classification problem we will need to complete a one hot encoding for all categorical variables

Y_train = keras.utils.to_categorical(Y_train,len(np.unique(infections)))
Y_test = keras.utils.to_categorical(Y_test,len(np.unique(infections)))
```

```python
# Defining and training a CNN

model=Sequential()
model.add(Conv2D(filters=16,kernel_size=2,padding="same",activation="relu",input_shape=(50,50,3)))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=32,kernel_size=2,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=64,kernel_size=2,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(500,activation="sigmoid"))
model.add(Dropout(0.2))
model.add(Dense(2,activation="sigmoid"))#2 represent output layer neurons 
model.summary()
```

```python
# compile the model with loss as categorical_crossentropy and using adam optimizer you can test result by trying RMSProp as well as Momentum
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Fit the model with min batch size as 50[can tune batch size to some factor of 2^power ] 
model.fit(X_train,Y_train,batch_size=50,epochs=20,verbose=1)
```

```python
accuracy = model.evaluate(X_test, Y_test, verbose=1)
print('\n', 'Test_Accuracy:-', accuracy[1])
```
